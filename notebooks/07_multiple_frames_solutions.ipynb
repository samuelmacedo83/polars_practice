```python
import polars as pl
from zipfile import ZipFile
```

**Exercise 61: Read all the files together from polars_sample.zip, and assign the list of frames to `data`**

```python
data = list([pl.read_csv(ZipFile('polars_sample.zip').open(i)) for i in ZipFile('polars_sample.zip').namelist()])
data
```

**Exercise 62: Assign the dataframes in data to individual frames `orders_jan`, `orders_feb`, `orders_mar`, `returns` and `users` respectively and display the head of each dataframe**

```python
orders_jan = data[2]
orders_jan.head()
```

```python
orders_feb = data[0]
orders_feb = orders_feb.with_column(pl.col('Sales').cast(pl.Float64, strict=False))
orders_feb.head()
```

```python
orders_mar = data[4]
orders_mar.head()
```

```python
returns = data[6]
returns.head()
```

```python
users = data[8]
users.head()
```

**Exercise 63: Append all rows of `orders_jan`, `orders_feb` and `orders_mar` into a single frame and assign it to `orders_all`**

```python
orders_all = pl.concat([orders_jan, orders_feb, orders_mar])
orders_all.head()
```

**Exercise 64: Join `orders_all` to `returns` on `Order ID` and assign it to `sales`**

```python
sales = returns.join(orders_all, on="Order ID")
sales.head()
```

**Exercise 65: Update `users` by joining `sales` to it on `Return ID` and `Category ID`**

```python
sales = users.join(sales, on=["Category ID", "Return ID"])
sales.head()
```

**Exercise 66: Display the Categories that had orders in `Jan` or `Mar`**

```python
pl.concat([orders_jan['Category'], orders_mar['Category']]).unique()
```

**Exercise 67: Display the Categories that had orders in each of the three months**

```python
set(orders_jan['Category']).intersection(set(orders_feb['Category'])).intersection(set(orders_mar['Category']))
```

**Exercise 68: Display the Categories that had orders only in `Jan` but not in `Feb` and `Mar`**

```python
set(orders_jan['Category']).difference(set(orders_feb['Category'])).difference(set(orders_mar['Category']))
```

**Exercise 69: Display the Categories that had orders in any of the three months but never together**

```python
set(orders_jan['Category']).symmetric_difference(set(orders_feb['Category'])).symmetric_difference(set(orders_mar['Category']))
```

**Exercise 70: One-hot encode the column `Region` in `users`, assign it to `ohe_region` and append it back to `users`**

```python
ohe_region = users['Region'].to_dummies()
ohe_region
```

```python
users = pl.concat([users, ohe_region], how="horizontal")
users
```
