```python
import polars as pl

data = pl.read_csv('polars_sample.csv', parse_dates = True)
data.drop(['age','category'])
data
```

**Exercise 71: Extract the `date` from `timestamp` and add it as a column `date` in `data`**

```python
data = data.with_column((pl.col('timestamp')).cast(pl.Date, strict=False).alias('date'))
data
```

**Exercise 72: Extract the `year`, `month`, `day` and `day-of-week` from `date` and add them as new columns `year`, `month`, `day` and `day_of_week` in `data`**

```python
data = data.with_column(pl.col("date").dt.strftime("%Y").cast(pl.Int64, strict=False).alias("year"))
data
```

```python
data = data.with_column(pl.col("date").dt.strftime("%m").cast(pl.Int64, strict=False).alias("month"))
data
```

```python
data = data.with_column(pl.col("date").dt.strftime("%d").cast(pl.Int64, strict=False).alias("day"))
data
```

```python
data = data.with_column(pl.col("date").dt.strftime("%w").cast(pl.Int64, strict=False).alias("day_of_week"))
data
```

**Exercise 73: Extract the `hour`, `minute` and `second` from `timestamp` and add them as new columns `hour`, `minute` and `second` in `data`**

```python
data = data.with_column(pl.col("timestamp").dt.strftime("%-H").cast(pl.Int64, strict=False).alias("hour"))
data
```

```python
data = data.with_column(pl.col("timestamp").dt.strftime("%-M").cast(pl.Int64, strict=False).alias("minute"))
data
```

```python
data = data.with_column(pl.col("timestamp").dt.strftime("%-S").cast(pl.Int64, strict=False).alias("second"))
data
```

**Exercise 74: Create a date column `date_new` in `data` using the `year`, `month` and `day` columns**

```python
data = data.with_columns([pl.date("year", "month", "day").dt.strftime("%Y-%m-%d").str.strptime(pl.Date, fmt="%Y-%m-%d").alias("date_new")])
data
```

**Exercise 75: Create a time column `time_new` in data using the `year`, `month`, `day`, `hour`, `minute` and `second` columns**

```python
data = data.with_columns([pl.datetime("year", "month", "day","hour","minute","second").dt.strftime("%Y-%m-%d %H:%M:%S").str.strptime(pl.Datetime, fmt="%Y-%m-%d %H:%M:%S").alias("time_new")])
data
```

**Exercise 76: Create a lead variable of `product` with shift of 2 called `next_product_2` in data**

```python
data = data.with_column(pl.col("product").shift(periods=-2).alias('next_product_2'))
data
```

**Exercise 77: Create a lag variable of `date` with single shift called `previous_date` in data**

```python
data = data.with_column(pl.col("date").shift(periods=1).alias('previous_date'))
data
```

**Exercise 78: Create a lag variable of `timestamp` by `user` with single shift called `previous_user_time` in data**

```python
data = data.sort(["user","timestamp"], reverse=False)
new_data = data.groupby("user", maintain_order=True).agg([pl.col("timestamp").shift(periods=1).alias('previous_user_time')]).explode(['previous_user_time']).drop('user')
data = pl.concat([data, new_data], how="horizontal")
data
```

**Exercise 79: Create a column `date_diff` which is the difference between `date` and `previous_date` in days**

```python
data = data.with_columns([(pl.col('date') - pl.col('previous_date')).alias('date_diff')])
data
```

**Exercise 80: Create a column `time_diff` which is the difference between `timestamp` and `previous_user_time` in seconds**

```python
data = data.with_columns([(pl.col('timestamp') - pl.col('previous_user_time')).alias('time_diff') / 1000])
data
```
