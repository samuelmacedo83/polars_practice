```python
import polars as pl
from sklearn.datasets import load_wine

data = pl.DataFrame(load_wine(as_frame=True).frame)
data
```

**Exercise 41: Calculate the sum of column `ash` in `data`**

```python
data['ash'].sum()
```

**Exercise 42: Calculate the average value of each column in `data`**

```python
data.mean()
```

**Exercise 43: Calculate the minimum value and maximum value of each column in `data` where `target` is 1**

```python
data.filter(pl.col('target') == 1).min()
```

```python
data.filter(pl.col('target') == 1).max()
```

**Exercise 44: Calculate the median value of columns `flavanoids`, `hue` and `magnesium` in `data`**

```python
data[['flavanoids', 'hue', 'magnesium']].median()
```

**Exercise 45: Calculate the mode of column `magnesium` in `data`**

```python
data.with_column(pl.col('magnesium').cast(pl.Int64, strict=False)).select([pl.col("magnesium").mode()])
```

**Exercise 46: Calculate the standard deviation, skewness and excess kurtosis of each column in `data`**

```python
data.std()
```

```python
data.select(pl.col('*').skew())
```

```python
data.select(pl.col('*').kurtosis())
```

**Exercise 47: Calculate the number of unique values in each column of data**

**Exercise 48: Calculate the covariance between columns `alcohol` and `color_intensity` in `data`**

```python
data.select(pl.cov("alcohol", "color_intensity"))
```

**Exercise 49: Calculate Pearson correlation and Spearman Rank correlation between columns `hue` and `color_intensity` in `data`**

```python
data.select(pl.pearson_corr("hue", "color_intensity"))
```

```python
data.select(pl.spearman_rank_corr("hue", "color_intensity"))
```

**Exercise 50: Calculate the correlation matrix between all pairs of columns in `data` and assign it to `corr_matrix`**

```python
corr_matrix = data.pearson_corr()
col_names = corr_matrix.columns
corr_matrix = corr_matrix.with_column(pl.Series(name="Columns", values=col_names))
corr_matrix = corr_matrix[['Columns'] + [ col for col in corr_matrix.columns if col != 'Columns' ]]
corr_matrix
```
