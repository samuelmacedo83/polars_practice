```python
import polars as pl

data = pl.read_csv('titanic.csv', null_values = '?')
data
```

**Exercise 31: Print all the column names and column types of `data` in `column-name : column-type` format**

```python
for i in range(data.width):
    print(f'{data.columns[i]} : {data.dtypes[i]}')
```

**Exercise 32: Convert the type of column `age` in `data` from `Float64` to `Float32`**

```python
data.with_column((pl.col('age')).cast(pl.Float32, strict=False))
```

**Exercise 33: Rename the column `home.dest` in `data` to `address`**

```python
data.names = data.rename({'home.dest': 'address'})
data.names
```

**Exercise 34: Sort `data` in descending order of `fare`**

```python
data= data.sort('fare', reverse=True)
data.head()
```

**Exercise 35: Create a column `deck` in `data` containing only the first character of `cabin`**

```python
data = data.with_column(pl.col('cabin').str.slice(0, length=1).alias('deck'))
data.head()
```

**Exercise 36: Create a column `family` in `data` which is the sum of two columns `sibsp` and `parch`**

```python
pl.Config.with_columns_kwargs = True
data = data.with_columns(family=pl.col('sibsp') + pl.col('parch'))
data.head()
```

**Exercise 37: Create a column `total_passengers` in `data` which is the sum of `family` and the passenger included and round the values of column `age` in `data` to the nearest integer**

```python
data = data.with_columns([( pl.col('family')+1).alias("total_passengers"),
                   (pl.col('age').round(0))])
data
```

**Exercise 38: Delete the used columns `sibsp` and `parch` and the first and last rows from `data`**

```python
data = data.drop(['sibsp', 'parch'])
data.head()
```

```python
data = data[1:1308, :]
data
```

**Exercise 39: Replace all string `null` values with `NA` as missing values in `data`**

```python
data = data.fill_null('NA')
data
```

**Exercise 40: Impute all the missing values in `data` with `-1` for numeric columns and `'unknown'` for string columns**

```python
data = data.fill_null(-1)
data
```

```python
data = data.with_column(pl.col(data.select(pl.col(pl.Utf8)).columns).str.replace_all("NA","unknown"))
data
```
